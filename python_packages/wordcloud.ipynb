```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def make_wordcloud(text):
    wordcloud = WordCloud(min_word_length=3).generate(text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```
