Actually Pillow - fork of Python Imaging Library (PIL) discontinued 2009

```python
from PIL import Image as PILImage
from IPython.display import display

with PILImage.open("data/flower.png") as img:
    img.thumbnail((300, 300))  # Resize image
    display(img)
```

```python
from PIL import ImageFilter
blurred_img = img.filter(ImageFilter.GaussianBlur(5))  # Apply Gaussian blur
display(blurred_img)
```

```python
from PIL import ImageDraw
draw_img = img.copy()
draw = ImageDraw.Draw(draw_img)
draw.rectangle([50, 150, 150, 250], outline="red")  # Draw a rectangle
draw.text((60, 160), "Flower", fill="white")  # Add text
display(draw_img)
```

```python
pixels = img.load()  # Load pixel data
for i in range(img.width):
    for j in range(img.height):
        r, g, b, a = pixels[i, j]
        pixels[i, j] = (r//2, g//2, b, a)  # Reduce red and green channels
display(img)
```
