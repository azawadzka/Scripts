## Compare extraction from text between OpenAI models and languages

````python
# generated by perplexity.ai: Tell me about recent quits from decisive positions in openai startup as short text, no bullet list
text = "OpenAI has recently experienced significant leadership changes, " \
    "with several high-profile departures raising concerns about the company's direction. " \
    "Notably, co-founder and chief scientist Ilya Sutskever left in May 2024, " \
    "followed closely by Jan Leike, a leader in the superalignment group. " \
    "Their exits were seen as part of a troubling trend, " \
    "as other key figures like Evan Morikawa and Peter Deng also announced their departures. " \
    "Most recently, co-founder John Schulman confirmed his move to rival AI startup Anthropic, " \
    "citing a desire to focus on AI alignment. " \
    "These resignations have led to speculation about internal disagreements over product direction " \
    "and the company's commitment to ethical AI practices, " \
    "particularly following a tumultuous period marked by CEO Sam Altman's brief ouster and subsequent reinstatement."
````

````python
# Basic retrieval

import instructor
from pydantic import BaseModel
from openai import OpenAI


class PersonInfo(BaseModel):
    name: str
    position: str
    date: str
    previous_company: str
    current_company: str

    def __str__(self):
        return ", ".join(f"{k}: {v}" for k, v in self.__dict__.items())


client = instructor.from_openai(OpenAI())

users_info = client.chat.completions.create_iterable(
    model="gpt-3.5-turbo",
    response_model=PersonInfo,
    messages=[{"role": "user", "content": text}],
)

for user_info in users_info:
    print(user_info)
    # name: Ilya Sutskever, position: Co-founder and Chief Scientist, date: May 2024, previous_company: OpenAI, current_company: N/A
    # name: Jan Leike, position: Leader in the Superalignment Group, date: N/A, previous_company: OpenAI, current_company: N/A
    # name: Evan Morikawa, position: N/A, date: N/A, previous_company: OpenAI, current_company: N/A
    # name: Peter Deng, position: N/A, date: N/A, previous_company: OpenAI, current_company: N/A
    # name: John Schulman, position: Co-founder, date: N/A, previous_company: OpenAI, current_company: Anthropic
````

## Parse instructions in NL to data structures
Hopefully useful for agentic design

````python
# What is the time difference between Lisbon and Buenos Aires?
# Prompted list of steps:
steps_in_nl = "1. Search Wikipedia page for current local time in Lisbon.\n2. Search Wikipedia page for current local time in Buenos Aires.\n3. Calculate the time difference between the two cities by subtracting the time in Lisbon from the time in Buenos Aires."
````

````python
from typing import Union
from pydantic import BaseModel

import instructor
from openai import OpenAI


class Step(BaseModel):
    step_number: int

class WikipediaStep(Step):
    wikipedia_page: str
    what_to_find: str

class CalculatorStep(Step):
    operation: str
    value1: str
    value2: str

STEP_TYPES = [WikipediaStep, CalculatorStep]

class StepsPlan(BaseModel):
    steps_list: list[Union[*STEP_TYPES]]

instructor_client = instructor.from_openai(OpenAI())

def create_plan(data: str) -> StepsPlan:
    return instructor_client.chat.completions.create(
        model="gpt-3.5-turbo",
        response_model=StepsPlan,
        messages=[
            {
                "role": "user",
                "content": f"Convert the following list of steps to a plan: {data}",
            },
        ],
    )

plan = create_plan(steps_in_nl)
plan
# StepsPlan(steps_list=[
#     WikipediaStep(step_number=1, wikipedia_page='Lisbon', what_to_find='current local time'), 
#     WikipediaStep(step_number=2, wikipedia_page='Buenos Aires', what_to_find='current local time'), 
#     CalculatorStep(step_number=3, operation='subtract', value1='time in Lisbon', value2='time in Buenos Aires')
#     ])
# issues: 
# - irregular responses
# - sometimes too broad searches, eg. wikipedia_page='current local time in Lisbon' instead of 'Lisbon'
````

````python
import instructor
from anthropic import Anthropic
from pydantic import BaseModel


class User(BaseModel):
    name: str
    age: int


client = instructor.from_anthropic(Anthropic())

# note that client.chat.completions.create will also work
resp = client.messages.create(
    model="claude-3-opus-20240229",
    max_tokens=1024,
    system="You are a world class AI that excels at extracting user data from a sentence",
    messages=[
        {
            "role": "user",
            "content": "Extract Jason is 25 years old.",
        }
    ],
    response_model=User,
)

assert isinstance(resp, User)
assert resp.name == "Jason"
assert resp.age == 25
````
